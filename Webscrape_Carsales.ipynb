```python
def extract(x, key1, key2):
    if key1 == '':
        start = 0
    else:
        start = x.find(key1) + len(key1)
    if key2 =='':
        end = len(x)
    else:
        end = x.find(key2)
    substring = x[start:end]
    return substring
```

```python
df['Make'] = df.apply(lambda row: extract(row['ModelMake'], '', '_'), axis=1)
df['Model'] = df.apply(lambda row: extract(row['ModelMake'], '_', ''), axis=1)
df['Make'] = df['Make'].str.replace(' ', '%20')
df['Model'] = df['Model'].str.replace(' ', '%20')
```

```python
def getflag(url2check, model, make):
    
    A = random.random()
    print(A*10)
    time.sleep(A*10)

    fp = urllib.request.urlopen(url2check)
    mybytes = fp.read()
    mystr = mybytes.decode("utf8")
    fp.close()
    soup = BeautifulSoup(mystr, 'html.parser')

    search = str(soup.find_all('div', attrs={"class":"navigation"})[0])
    delimiter = '{"identifier'
    search = delimiter+search #so that first element is the string before first occurrence
    searchcut = ['{}{}'.format(delimiter, s) for s in search.split(delimiter) if s]
    searchcut = searchcut[1:len(searchcut)]
    count_art = len(searchcut) #number of articles on first page with that keyword

    if len(searchcut) > 0:
        headlines = [extract(x, "headline", "subHeading") for x in searchcut]
        keywordcount = [x.replace("-", "").lower().count(model.lower()) for x in headlines] #count how many times the article is actually about model
        dates = [extract(x, 'dateAvailable":', '') for x in searchcut] #to check recency
        dates = [extract(x, '', '}') for x in dates]
        yeardates = [int(''.join(filter(str.isdigit, x))) for x in dates] #just year for simplicity

        #function for recent articles that are listed as ex. January 31st, not January 2022
        for i in range(len(yeardates)-1):
            if yeardates[i] < 100:
                yeardates[i] = 2022
            else:
                yeardates[i] = yeardates[i]

        avgyear = str(round(sum(yeardates)/len(searchcut), 2))
        if sum(keywordcount) <= 1:
            flag = "1" #ie possibly manually check
        else:
            flag = "2" #should be good
    else:
        flag = "0" #no data
        avgyear = "0"
        
    return flag+"_"+avgyear+"_"+str(count_art)
```
